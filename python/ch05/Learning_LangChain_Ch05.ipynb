# 2. State (ìƒíƒœ) ì •ì˜

ì±—ë´‡ì´ ê´€ë¦¬í•  ë°ì´í„° êµ¬ì¡°ë¥¼ ì •ì˜í•©ë‹ˆë‹¤.

````python
from typing import Annotated, TypedDict
from langgraph.graph import add_messages

class State(TypedDict):
    """
    ì±—ë´‡ì˜ ìƒíƒœë¥¼ ì •ì˜í•©ë‹ˆë‹¤.
    
    messages: ëŒ€í™” ê¸°ë¡ ë¦¬ìŠ¤íŠ¸
    - Annotated[list, add_messages]: 
      ìƒˆ ë©”ì‹œì§€ê°€ ì˜¤ë©´ ë®ì–´ì“°ì§€ ì•Šê³  ì¶”ê°€í•¨
    """
    messages: Annotated[list, add_messages]

print("âœ… State ì •ì˜ ì™„ë£Œ")
print("   messages: ëŒ€í™” ê¸°ë¡ì„ ì €ì¥í•˜ëŠ” ë¦¬ìŠ¤íŠ¸")
````

# 3. Node (ë…¸ë“œ) ì •ì˜

LLMì„ í˜¸ì¶œí•˜ëŠ” ì±—ë´‡ ë…¸ë“œë¥¼ ë§Œë“­ë‹ˆë‹¤.

````python
from langchain_ollama import ChatOllama

# LLM ëª¨ë¸
model = ChatOllama(model='llama3.2')

def chatbot(state: State):
    """
    ì±—ë´‡ ë…¸ë“œ
    
    1. í˜„ì¬ ìƒíƒœì—ì„œ ë©”ì‹œì§€ ê¸°ë¡ì„ ê°€ì ¸ì˜´
    2. LLMì—ê²Œ ì „ë‹¬í•˜ì—¬ ì‘ë‹µ ìƒì„±
    3. ì‘ë‹µì„ ë©”ì‹œì§€ ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€í•˜ì—¬ ë°˜í™˜
    """
    # LLM í˜¸ì¶œ
    answer = model.invoke(state['messages'])
    
    # ì‘ë‹µì„ ë©”ì‹œì§€ì— ì¶”ê°€
    return {'messages': [answer]}

print("âœ… chatbot ë…¸ë“œ ì •ì˜ ì™„ë£Œ")
````

# 4. Graph (ê·¸ë˜í”„) êµ¬ì„±

ë…¸ë“œì™€ ì—£ì§€ë¥¼ ì—°ê²°í•˜ì—¬ ê·¸ë˜í”„ë¥¼ ì™„ì„±í•©ë‹ˆë‹¤.

````python
from langgraph.graph import StateGraph, START, END

# ê·¸ë˜í”„ ë¹Œë” ìƒì„±
builder = StateGraph(State)

# ë…¸ë“œ ì¶”ê°€
# ì²« ë²ˆì§¸ ì¸ì: ë…¸ë“œ ì´ë¦„ (ê³ ìœ í•´ì•¼ í•¨)
# ë‘ ë²ˆì§¸ ì¸ì: ì‹¤í–‰í•  í•¨ìˆ˜
builder.add_node('chatbot', chatbot)

# ì—£ì§€ ì¶”ê°€
builder.add_edge(START, 'chatbot')  # ì‹œì‘ â†’ ì±—ë´‡
builder.add_edge('chatbot', END)     # ì±—ë´‡ â†’ ì¢…ë£Œ

# ê·¸ë˜í”„ ì»´íŒŒì¼
graph = builder.compile()

print("âœ… ê·¸ë˜í”„ ì»´íŒŒì¼ ì™„ë£Œ")
````

## stream vs invoke

| ë©”ì„œë“œ | ë™ì‘ | ì‚¬ìš© ì‹œë‚˜ë¦¬ì˜¤ |
|--------|------|---------------|
| **invoke** | ìµœì¢… ê²°ê³¼ë§Œ ë°˜í™˜ | ê²°ê³¼ë§Œ í•„ìš”í•  ë•Œ |
| **stream** | ë‹¨ê³„ë³„ ê²°ê³¼ ë°˜í™˜ | ì§„í–‰ ìƒí™©ì„ ë³´ê³  ì‹¶ì„ ë•Œ |

````python
from langchain_core.messages import HumanMessage

# ì…ë ¥ ì¤€ë¹„
input_data = {'messages': [HumanMessage(content='ì•ˆë…•í•˜ì„¸ìš”!')]}

print("=== streamìœ¼ë¡œ ì‹¤í–‰ ===")
print(f"ì…ë ¥: {input_data['messages'][0].content}\n")

# streamìœ¼ë¡œ ì‹¤í–‰ (ë‹¨ê³„ë³„ ì¶œë ¥)
for chunk in graph.stream(input_data):
    print(f"chunk: {chunk}")
````

````python
# invokeë¡œ ì‹¤í–‰ (ìµœì¢… ê²°ê³¼ë§Œ)
result = graph.invoke(input_data)

print("\n=== invokeë¡œ ì‹¤í–‰ ===")
print(f"ì‘ë‹µ: {result['messages'][-1].content}")
````

# 2. ë‘ ê°€ì§€ LLM ì„¤ì •

````python
from langchain_ollama import ChatOllama

# SQL ì¿¼ë¦¬ ìƒì„±ìš© (ì •í™•ì„± ì¤‘ìš” â†’ ë‚®ì€ temperature)
model_low_temp = ChatOllama(model='llama3.2', temperature=0.1)

# ìì—°ì–´ ì„¤ëª… ìƒì„±ìš© (ìì—°ìŠ¤ëŸ¬ì›€ ì¤‘ìš” â†’ ë†’ì€ temperature)
model_high_temp = ChatOllama(model='llama3.2', temperature=0.7)

print("âœ… ë‘ ê°€ì§€ LLM ì„¤ì • ì™„ë£Œ")
print("   - model_low_temp (temperature=0.1): SQL ìƒì„±ìš©")
print("   - model_high_temp (temperature=0.7): ì„¤ëª… ìƒì„±ìš©")
````

# 3. State ì •ì˜ (í™•ì¥ëœ ë²„ì „)

ì´ë²ˆì—ëŠ” ë” ë§ì€ ë°ì´í„°ë¥¼ ì €ì¥í•©ë‹ˆë‹¤.

````python
from typing import Annotated, TypedDict
from langgraph.graph.message import add_messages

class State(TypedDict):
    """
    SQL ìƒì„±ê¸°ì˜ ìƒíƒœ
    
    messages: ëŒ€í™” ê¸°ë¡
    user_query: ì‚¬ìš©ìì˜ ì›ë³¸ ì§ˆë¬¸
    sql_query: ìƒì„±ëœ SQL ì¿¼ë¦¬
    sql_explanation: SQLì— ëŒ€í•œ ì„¤ëª…
    """
    messages: Annotated[list, add_messages]  # ëŒ€í™” ê¸°ë¡
    user_query: str                          # ì…ë ¥
    sql_query: str                           # ì¶œë ¥ 1
    sql_explanation: str                     # ì¶œë ¥ 2

# ì…ë ¥ê³¼ ì¶œë ¥ íƒ€ì… ì •ì˜ (ì„ íƒ ì‚¬í•­ì´ì§€ë§Œ ëª…í™•ì„±ì„ ìœ„í•´)
class Input(TypedDict):
    user_query: str

class Output(TypedDict):
    sql_query: str
    sql_explanation: str

print("âœ… State, Input, Output ì •ì˜ ì™„ë£Œ")
````

# 4. ë…¸ë“œ 1: SQL ì¿¼ë¦¬ ìƒì„±

ì‚¬ìš©ìì˜ ìì—°ì–´ ì§ˆë¬¸ì„ SQL ì¿¼ë¦¬ë¡œ ë³€í™˜í•©ë‹ˆë‹¤.

````python
from langchain_core.messages import HumanMessage, SystemMessage

# SQL ìƒì„±ìš© ì‹œìŠ¤í…œ í”„ë¡¬í”„íŠ¸
generate_prompt = SystemMessage(
    content='ë‹¹ì‹ ì€ ì¹œì ˆí•œ ë°ì´í„° ë¶„ì„ê°€ì…ë‹ˆë‹¤. ì‚¬ìš©ìì˜ ì§ˆë¬¸ì„ ë°”íƒ•ìœ¼ë¡œ SQL ì¿¼ë¦¬ë¥¼ ì‘ì„±í•˜ì„¸ìš”.'
)

def generate_sql(state: State) -> State:
    """
    ë…¸ë“œ 1: SQL ì¿¼ë¦¬ ìƒì„±
    
    1. ì‚¬ìš©ì ì§ˆë¬¸ì„ HumanMessageë¡œ ë³€í™˜
    2. ì‹œìŠ¤í…œ í”„ë¡¬í”„íŠ¸ + ëŒ€í™” ê¸°ë¡ + ì‚¬ìš©ì ì§ˆë¬¸ ì¡°í•©
    3. LLMìœ¼ë¡œ SQL ì¿¼ë¦¬ ìƒì„±
    4. ê²°ê³¼ë¥¼ Stateì— ì €ì¥
    """
    user_message = HumanMessage(content=state['user_query'])
    messages = [generate_prompt, *state['messages'], user_message]
    
    # ë‚®ì€ temperatureë¡œ ì •í™•í•œ SQL ìƒì„±
    res = model_low_temp.invoke(messages)
    
    return {
        'sql_query': res.content,
        'messages': [user_message, res],  # ëŒ€í™” ê¸°ë¡ ì—…ë°ì´íŠ¸
    }

print("âœ… generate_sql ë…¸ë“œ ì •ì˜ ì™„ë£Œ")
````

# 5. ë…¸ë“œ 2: SQL ì„¤ëª… ìƒì„±

ìƒì„±ëœ SQL ì¿¼ë¦¬ë¥¼ ì‚¬ìš©ìê°€ ì´í•´í•˜ê¸° ì‰½ê²Œ ì„¤ëª…í•©ë‹ˆë‹¤.

````python
# SQL ì„¤ëª…ìš© ì‹œìŠ¤í…œ í”„ë¡¬í”„íŠ¸
explain_prompt = SystemMessage(
    content='ë‹¹ì‹ ì€ ì¹œì ˆí•œ ë°ì´í„° ë¶„ì„ê°€ì…ë‹ˆë‹¤. ì‚¬ìš©ìì—ê²Œ SQL ì¿¼ë¦¬ë¥¼ ì„¤ëª…í•˜ì„¸ìš”.'
)

def explain_sql(state: State) -> State:
    """
    ë…¸ë“œ 2: SQL ì¿¼ë¦¬ ì„¤ëª…
    
    1. ì´ì „ ëŒ€í™” ê¸°ë¡ (ì‚¬ìš©ì ì§ˆë¬¸ + SQL ì¿¼ë¦¬) ê°€ì ¸ì˜¤ê¸°
    2. LLMìœ¼ë¡œ ìì—°ìŠ¤ëŸ¬ìš´ ì„¤ëª… ìƒì„±
    3. ê²°ê³¼ë¥¼ Stateì— ì €ì¥
    """
    messages = [
        explain_prompt,
        *state['messages'],  # ì´ì „ ë‹¨ê³„ì˜ ì‚¬ìš©ì ì§ˆë¬¸ê³¼ SQL ì¿¼ë¦¬
    ]
    
    # ë†’ì€ temperatureë¡œ ìì—°ìŠ¤ëŸ¬ìš´ ì„¤ëª… ìƒì„±
    res = model_high_temp.invoke(messages)
    
    return {
        'sql_explanation': res.content,
        'messages': res,  # ëŒ€í™” ê¸°ë¡ ì—…ë°ì´íŠ¸
    }

print("âœ… explain_sql ë…¸ë“œ ì •ì˜ ì™„ë£Œ")
````

# 6. ê·¸ë˜í”„ êµ¬ì„±

ë‘ ë…¸ë“œë¥¼ ìˆœì°¨ì ìœ¼ë¡œ ì—°ê²°í•©ë‹ˆë‹¤.

```
START â†’ generate_sql â†’ explain_sql â†’ END
```

````python
from langgraph.graph import StateGraph, START, END

# ê·¸ë˜í”„ ë¹Œë” ìƒì„± (ì…ë ¥/ì¶œë ¥ íƒ€ì… ì§€ì •)
builder = StateGraph(State, input=Input, output=Output)

# ë…¸ë“œ ì¶”ê°€
builder.add_node('generate_sql', generate_sql)
builder.add_node('explain_sql', explain_sql)

# ì—£ì§€ ì¶”ê°€ (ìˆœì°¨ ì‹¤í–‰)
builder.add_edge(START, 'generate_sql')       # ì‹œì‘ â†’ SQL ìƒì„±
builder.add_edge('generate_sql', 'explain_sql')  # SQL ìƒì„± â†’ SQL ì„¤ëª…
builder.add_edge('explain_sql', END)          # SQL ì„¤ëª… â†’ ì¢…ë£Œ

# ê·¸ë˜í”„ ì»´íŒŒì¼
graph = builder.compile()

print("âœ… ê·¸ë˜í”„ ì»´íŒŒì¼ ì™„ë£Œ")
````

````python
# ê·¸ë˜í”„ êµ¬ì¡° ì‹œê°í™”
print("=== ê·¸ë˜í”„ êµ¬ì¡° ===")
print(graph.get_graph().draw_mermaid())
````

# 3. ìƒ˜í”Œ ë¬¸ì„œ ë° ë²¡í„° ì €ì¥ì†Œ ì¤€ë¹„

ë‘ ê°€ì§€ ë„ë©”ì¸ì˜ ë¬¸ì„œë¥¼ ì¤€ë¹„í•©ë‹ˆë‹¤.

````python
from langchain_core.documents import Document
from langchain_core.vectorstores.in_memory import InMemoryVectorStore

# ìƒ˜í”Œ ë¬¸ì„œ
sample_docs = [
    Document(
        page_content='''# í™˜ì ì˜ë£Œ ê¸°ë¡
- ì´ë¦„: í™ê¸¸ë™
- ì„±ë³„: ë‚¨
- ë‚˜ì´: 30ì„¸
## ì§„ë£Œ ë‚´ì—­
- 2021ë…„ 1ì›” 13ì¼: ê°ê¸°ë¡œ ì¸í•œ ë°œì—´ë¡œ ë³‘ì› ë°©ë¬¸
- 2022ë…„ 3ì›” 15ì¼: ë¹„ì—¼ ì§„ë‹¨ìœ¼ë¡œ ì•½ ì²˜ë°©
- 2022ë…„ 5ì›” 20ì¼: í”¼ë¶€ê³¼ ì§„ë£Œ ë° ì•½ ì²˜ë°©
- 2022ë…„ 6ì›” 10ì¼: ë°œì—´ ë° ê¸°ì¹¨ìœ¼ë¡œ ì½”ë¡œë‚˜ ê²€ì‚¬ ë° ìŒì„± íŒì •
- 2022ë…„ 7ì›” 2ì¼: ì½”ë¡œë‚˜ ì§„ë‹¨ ê²€ì‚¬ ë° ì–‘ì„± íŒì •
        ''',
        metadata={'domain': 'records'}
    ),
    Document(
        page_content='''# ë³´í—˜ FAQ
- Q: ê³¼ê±°ì— ë³‘ë ¥ì´ ìˆëŠ”ë° ê°€ì…ê°€ëŠ¥í•œê°€ìš”?
- A: ì¹˜ë£Œê¸°ê°„, í˜„ì¬ ìƒíƒœ, í›„ìœ ì¦ ì—¬ë¶€, ì—°ë ¹ ë“±ì— ë”°ë¼ ë‹¤ë¥´ë¯€ë¡œ ê°€ì… ê°€ëŠ¥ ì—¬ë¶€ëŠ” ê°€ê¹Œìš´ ëŒ€ë¦¬ì ì—ì„œ ìƒë‹´ ë°›ìœ¼ì„¸ìš”.

- Q: ë³´í—˜ê¸ˆ ì²­êµ¬ ë°©ë²•ì€ ì–´ë–»ê²Œ ë˜ë‚˜ìš”??
- A: ë³´í—˜ê¸ˆ ì²­êµ¬ëŠ” ë³´í—˜ê¸ˆ ì²­êµ¬ì„œ ì‘ì„± í›„ ê°€ê¹Œìš´ ëŒ€ë¦¬ì ì„ í†µí•´ ì§„í–‰ ê°€ëŠ¥í•©ë‹ˆë‹¤.

- Q: ì½”ë¡œë‚˜ 19ë„ ë³´í—˜ ì ìš©ì´ ë˜ë‚˜ìš”?
- A: ì½”ë¡œë‚˜ 19ëŠ” ë³´í—˜ ëŒ€ìƒì— í¬í•¨ë˜ì§€ ì•ŠìŠµë‹ˆë‹¤.

- Q: ë³´í—˜ê¸ˆ ì§€ê¸‰ì´ ê±°ë¶€ë˜ëŠ” ê²½ìš°ëŠ” ì–´ë–¤ ê²½ìš°ì¸ê°€ìš”?
- A: ë³´í—˜ê¸ˆ ì§€ê¸‰ ê±°ë¶€ ì‚¬ìœ ëŠ” ë³´í—˜ ê³„ì•½ì„œì— ëª…ì‹œë˜ì–´ ìˆìŠµë‹ˆë‹¤. ìì„¸í•œ ì‚¬í•­ì€ ê³„ì•½ì„œë¥¼ ì°¸ì¡°í•˜ì„¸ìš”
        ''',
        metadata={'domain': 'insurance'}
    ),
]

print("=== ìƒ˜í”Œ ë¬¸ì„œ ===")
for doc in sample_docs:
    print(f"\në„ë©”ì¸: {doc.metadata['domain']}")
    print(f"ë‚´ìš©: {doc.page_content[:100]}...")
````

````python
# ë²¡í„° ì €ì¥ì†Œ ìƒì„± (ë„ë©”ì¸ë³„)
medical_records_store = InMemoryVectorStore.from_documents(sample_docs, embeddings)
medical_records_retriever = medical_records_store.as_retriever()

insurance_faqs_store = InMemoryVectorStore.from_documents(sample_docs, embeddings)
insurance_faqs_retriever = insurance_faqs_store.as_retriever()

print("âœ… ë²¡í„° ì €ì¥ì†Œ ì¤€ë¹„ ì™„ë£Œ")
print("   - medical_records_retriever: ì˜ë£Œ ê¸°ë¡ ê²€ìƒ‰")
print("   - insurance_faqs_retriever: ë³´í—˜ FAQ ê²€ìƒ‰")
````

# 4. State ì •ì˜

````python
from typing import Annotated, Literal, TypedDict
from langgraph.graph.message import add_messages

class State(TypedDict):
    """
    Multi-RAG ì‹œìŠ¤í…œì˜ ìƒíƒœ
    """
    messages: Annotated[list, add_messages]    # ëŒ€í™” ê¸°ë¡
    user_query: str                            # ì‚¬ìš©ì ì§ˆë¬¸
    domain: Literal['records', 'insurance']    # ë¶„ë¥˜ëœ ë„ë©”ì¸
    documents: list[Document]                  # ê²€ìƒ‰ëœ ë¬¸ì„œ
    answer: str                                # ìµœì¢… ë‹µë³€

class Input(TypedDict):
    user_query: str

class Output(TypedDict):
    documents: list[Document]
    answer: str

print("âœ… State ì •ì˜ ì™„ë£Œ")
````

# 5. ë…¸ë“œ ì •ì˜

## ë…¸ë“œ 1: Router (ë„ë©”ì¸ ë¶„ë¥˜)

````python
from langchain_core.messages import HumanMessage, SystemMessage

# ë¼ìš°í„° í”„ë¡¬í”„íŠ¸
router_prompt = SystemMessage(
    content='''ì‚¬ìš©ì ë¬¸ì˜ë¥¼ ì–´ëŠ ë„ë©”ì¸ìœ¼ë¡œ ë¼ìš°íŒ…í• ì§€ ê²°ì •í•˜ì„¸ìš”. ì„ íƒí•  ìˆ˜ ìˆëŠ” ë‘ ê°€ì§€ ë„ë©”ì¸ì€ ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤.
- records: ì§„ë‹¨, ì¹˜ë£Œ, ì²˜ë°©ê³¼ ê°™ì€ í™˜ìì˜ ì˜ë£Œ ê¸°ë¡ì„ í¬í•¨í•©ë‹ˆë‹¤.  
- insurance: ë³´í—˜ ì •ì±…, ì²­êµ¬, ë³´ì¥ì— ëŒ€í•œ ìì£¼ ë¬»ëŠ” ì§ˆë¬¸ì„ í¬í•¨í•©ë‹ˆë‹¤.  

ë„ë©”ì¸ ì´ë¦„ë§Œ ì¶œë ¥í•˜ì„¸ìš”.'''
)

def router_node(state: State) -> State:
    """
    Router ë…¸ë“œ: ì§ˆë¬¸ì˜ ë„ë©”ì¸ì„ ë¶„ë¥˜
    
    ì¶œë ¥: 'records' ë˜ëŠ” 'insurance'
    """
    user_message = HumanMessage(content=state['user_query'])
    messages = [router_prompt, *state['messages'], user_message]
    
    res = model_low_temp.invoke(messages)
    
    # ë„ë©”ì¸ ê²°ì • (ì¶œë ¥ì—ì„œ ë„ë©”ì¸ ì¶”ì¶œ)
    domain = 'records' if 'records' in res.content.lower() else 'insurance'
    
    print(f"ğŸ”€ Router: '{state['user_query'][:30]}...' â†’ {domain}")
    
    return {
        'domain': domain,
        'messages': [user_message, res],
    }

print("âœ… router_node ì •ì˜ ì™„ë£Œ")
````

## ì¡°ê±´ë¶€ ë¶„ê¸° í•¨ìˆ˜

````python
def pick_retriever(state: State) -> Literal['retrieve_medical_records', 'retrieve_insurance_faqs']:
    """
    ì¡°ê±´ë¶€ ë¶„ê¸° í•¨ìˆ˜
    
    ë„ë©”ì¸ì— ë”°ë¼ ë‹¤ìŒ ë…¸ë“œë¥¼ ê²°ì •:
    - records â†’ retrieve_medical_records
    - insurance â†’ retrieve_insurance_faqs
    """
    if state['domain'] == 'records':
        return 'retrieve_medical_records'
    else:
        return 'retrieve_insurance_faqs'

print("âœ… pick_retriever ì¡°ê±´ë¶€ ë¶„ê¸° í•¨ìˆ˜ ì •ì˜ ì™„ë£Œ")
````

## ë…¸ë“œ 2a, 2b: ê²€ìƒ‰ ë…¸ë“œë“¤

````python
def retrieve_medical_records(state: State) -> State:
    """
    ì˜ë£Œ ê¸°ë¡ ê²€ìƒ‰ ë…¸ë“œ
    """
    documents = medical_records_retriever.invoke(state['user_query'])
    print(f"ğŸ¥ ì˜ë£Œ ê¸°ë¡ ê²€ìƒ‰: {len(documents)}ê°œ ë¬¸ì„œ ì°¾ìŒ")
    return {'documents': documents}

def retrieve_insurance_faqs(state: State) -> State:
    """
    ë³´í—˜ FAQ ê²€ìƒ‰ ë…¸ë“œ
    """
    documents = insurance_faqs_retriever.invoke(state['user_query'])
    print(f"ğŸ“‹ ë³´í—˜ FAQ ê²€ìƒ‰: {len(documents)}ê°œ ë¬¸ì„œ ì°¾ìŒ")
    return {'documents': documents}

print("âœ… ê²€ìƒ‰ ë…¸ë“œë“¤ ì •ì˜ ì™„ë£Œ")
````

## ë…¸ë“œ 3: ë‹µë³€ ìƒì„±

````python
# ë„ë©”ì¸ë³„ í”„ë¡¬í”„íŠ¸
medical_records_prompt = SystemMessage(
    content='ë‹¹ì‹ ì€ ìœ ëŠ¥í•œ ì˜ë£Œ ì±—ë´‡ì…ë‹ˆë‹¤. ì§„ë‹¨, ì¹˜ë£Œ, ì²˜ë°©ê³¼ ê°™ì€ í™˜ìì˜ ì˜ë£Œ ê¸°ë¡ì„ ê¸°ë°˜ìœ¼ë¡œ ì§ˆë¬¸ì— ë‹µí•˜ì„¸ìš”.'
)

insurance_faqs_prompt = SystemMessage(
    content='ë‹¹ì‹ ì€ ìœ ëŠ¥í•œ ì˜ë£Œ ë³´í—˜ ì±—ë´‡ì…ë‹ˆë‹¤. ë³´í—˜ ì •ì±…, ì²­êµ¬ ë° ë³´ì¥ì— ëŒ€í•œ ìì£¼ ë¬»ëŠ” ì§ˆë¬¸ì— ë‹µí•˜ì„¸ìš”.'
)

def generate_answer(state: State) -> State:
    """
    ë‹µë³€ ìƒì„± ë…¸ë“œ
    
    ë„ë©”ì¸ì— ë§ëŠ” í”„ë¡¬í”„íŠ¸ë¥¼ ì‚¬ìš©í•˜ì—¬ ë‹µë³€ ìƒì„±
    """
    # ë„ë©”ì¸ì— ë”°ë¥¸ í”„ë¡¬í”„íŠ¸ ì„ íƒ
    if state['domain'] == 'records':
        prompt = medical_records_prompt
    else:
        prompt = insurance_faqs_prompt
    
    messages = [
        prompt,
        *state['messages'],
        HumanMessage(content=f'Documents: {state["documents"]}'),
    ]
    
    res = model_high_temp.invoke(messages)
    
    return {
        'answer': res.content,
        'messages': res,
    }

print("âœ… generate_answer ë…¸ë“œ ì •ì˜ ì™„ë£Œ")
````

# 6. ê·¸ë˜í”„ êµ¬ì„± (ì¡°ê±´ë¶€ ë¶„ê¸° í¬í•¨)

````python
from langgraph.graph import StateGraph, START, END

# ê·¸ë˜í”„ ë¹Œë”
builder = StateGraph(State, input=Input, output=Output)

# ë…¸ë“œ ì¶”ê°€
builder.add_node('router', router_node)
builder.add_node('retrieve_medical_records', retrieve_medical_records)
builder.add_node('retrieve_insurance_faqs', retrieve_insurance_faqs)
builder.add_node('generate_answer', generate_answer)

# ì—£ì§€ ì¶”ê°€
builder.add_edge(START, 'router')

# â­ ì¡°ê±´ë¶€ ì—£ì§€ ì¶”ê°€
builder.add_conditional_edges('router', pick_retriever)

# ê²€ìƒ‰ ë…¸ë“œë“¤ â†’ ë‹µë³€ ìƒì„±
builder.add_edge('retrieve_medical_records', 'generate_answer')
builder.add_edge('retrieve_insurance_faqs', 'generate_answer')

# ë‹µë³€ ìƒì„± â†’ ì¢…ë£Œ
builder.add_edge('generate_answer', END)

# ê·¸ë˜í”„ ì»´íŒŒì¼
graph = builder.compile()

print("âœ… ê·¸ë˜í”„ ì»´íŒŒì¼ ì™„ë£Œ (ì¡°ê±´ë¶€ ë¶„ê¸° í¬í•¨)")
````

````python
# ê·¸ë˜í”„ êµ¬ì¡° ì‹œê°í™”
print("=== ê·¸ë˜í”„ êµ¬ì¡° ===")
print(graph.get_graph().draw_mermaid())
````
